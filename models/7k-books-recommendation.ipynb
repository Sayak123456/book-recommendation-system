```python
final_df['authors']=final_df['authors'].apply(lambda x:x.replace(" ","").split())
final_df['categories']=final_df['categories'].apply(lambda x:x.replace(" ",",").split())
final_df['description']=final_df['description'].apply(lambda x:x.replace(",",""))
final_df['description']=final_df['description'].apply(lambda x:x.replace(" ",",").split())
final_df['published_year']=final_df['published_year'].apply(lambda x:x.split())
final_df['average_rating']=final_df['average_rating'].apply(lambda x:x.split())
```

```python
final_df['tags']=final_df['authors']+final_df['categories']+final_df['description']+final_df['published_year']+final_df['average_rating']
```

```python
final_df.head()
```

```python
final_df=final_df.drop(columns=['authors','categories','description','published_year','average_rating'])
```

```python
final_df.head()
```

```python
final_df['tags']=final_df['tags'].apply(lambda x:" ".join(x))
```

```python
final_df.sample(5)
```

```python
x=final_df['tags'].values.tolist()
y=''
for i in x:
    y=y+i
len(y)
```

```python
len(set(y.split()))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(final_df['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity[0]
```

```python
def recommend(books):
    index=final_df[final_df['title']==books].index[0]
    distance=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distance[1:6]:
        print(final_df.iloc[i[0]].title)
```

```python
recommend('Data Structures and Algorithms in Java')
```

```python
recommend('The Lord of the Rings')
```
